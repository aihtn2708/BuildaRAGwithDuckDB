```python
df = duckdb.sql("CREATE TABLE SALES AS SELECT * FROM 'sales_data.csv'")
```

```python
duckdb.sql("SELECT  FROM SALES").df()
```

```python
top10customers = duckdb.sql("""SELECT "Customer Name",
                                     SUM(CAST("Sales Amount (USD)" as INTEGER)) AS total_sales
                                     FROM SALES
                                     GROUP BY "Customer Name"
                                     ORDER BY total_sales DESC
                                     LIMIT 10""")
```

```python
top10customers.df()
```

```python
RevenueByChannel = duckdb.sql("""SELECT "Sales Channel",
                                     SUM(CAST("Sales Amount (USD)" as INTEGER)) AS total_sales
                                     FROM SALES
                                     GROUP BY "Sales Channel"
                                     ORDER BY total_sales DESC""")
RevenueByChannel.df()
```

```python
customerFeedbackStats = duckdb.sql("""SELECT "Customer Feedback",
                                     COUNT(*) AS feedback_count
                                     FROM SALES
                                     GROUP BY "Customer Feedback"
                                     ORDER BY feedback_count DESC""")
customerFeedbackStats.df()
```
